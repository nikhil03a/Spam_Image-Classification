```python
import os
import random
import cv2
import numpy as np
from PIL import Image,ImageSequence
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
output_directory = 'augmented_images2'
```

```python
image_directory = 'personal_image_spam/'
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
list_images = os.listdir(image_directory)
for image_file in list_images:
    try:
        if image_file.lower().endswith('.gif'):
            gif_path = os.path.join(image_directory, image_file)
            gif = Image.open(gif_path)
            frames = [frame.copy() for frame in ImageSequence.Iterator(gif)]
            for frame in frames:
                frame = frame.resize((32, 32))
                image_array = img_to_array(frame)
                if image_array.shape[2] == 1:
                    image_array = cv2.cvtColor(image_array, cv2.COLOR_GRAY2RGB)
                image_array = image_array.reshape((1,) + image_array.shape)
                print(image_array.shape)
                i = 0
                for batch in datagen.flow(image_array, batch_size=1, save_to_dir=output_directory, save_prefix='spm', save_format='jpeg'):
                    i += 1
                    if i > 4:
                        break
        else:
            image = Image.open(os.path.join(image_directory, image_file))
            image = image.resize((32, 32))
            image_array = img_to_array(image)
            if image_array.shape[2] == 1:
                image_array = cv2.cvtColor(image_array, cv2.COLOR_GRAY2RGB)
            image_array = image_array.reshape((1,) + image_array.shape)
            i = 0
            print(image_array.shape)
            i = 0
            for batch in datagen.flow(image_array, batch_size=1, save_to_dir=output_directory, save_prefix='spm', save_format='jpeg'):
                i += 1
                if i > 4:
                    break
    except Exception as e:
        print(f"Error processing {image_file}: {e}")
```

```python
train_data = []
train_labels = []
test_data = []
test_labels = []
class_names = np.array([0,1])
images = os.listdir(output_directory)
train_len = int(len(images)*0.75)
random.shuffle(images)
for i in range(len(images)):
    try:
        image = img_to_array(Image.open(os.path.join(output_directory, images[i])))
        if i < train_len:
            train_data.append(image)
            if images[i].lower().startswith("spm"):
                train_labels.append(1)
            else:
                train_labels.append(0)
        else:
            test_data.append(image)
            if images[i].lower().startswith("spm"):
                test_labels.append(1)
            else:
                test_labels.append(0)
    except Exception as e:
        print(f"Error processing {images[i]}: {e}")        
```

```python
cnn = models.Sequential([
    #cnn
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    #ann
    layers.Flatten(),
    layers.Dense(100,activation='relu'),
    layers.Dense(2,activation='softmax')
])

cnn.compile(optimizer='adam',
           loss='sparse_categorical_crossentropy',
           metrics=['accuracy']
)
cnn.fit(X_train,y_train,epochs=10)
```
